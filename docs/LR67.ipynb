```python
# Useful libraries
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
```

```python
# Data loading
filename='../LR67/bsa_waxs/bsa.dat'
filename_out='../LR67/bsa_waxs/bsa_fmt.dat'
data = genfromtxt(filename)
```

```python
np.savetxt(filename_out,data, fmt='%.18e', delimiter=' ')
```

```python
plt.figure(figsize=(18, 12), dpi= 80, facecolor='w', edgecolor='k')
plt.subplot(211)
plt.grid()
plt.xlabel('q (inverse Angstroem)')
plt.ylabel('intensity')
plt.ylim(-1e5,3e5)
plt.plot(data[:,0],data[:,1],'kX')
plt.subplot(212)
plt.grid()
plt.xlabel('q (inverse Angstroem)')
plt.ylabel('intensity')
#plt.ylim(1e5,1e10)
plt.semilogy(data[:,0],data[:,1],'kX')
plt.show()
```
