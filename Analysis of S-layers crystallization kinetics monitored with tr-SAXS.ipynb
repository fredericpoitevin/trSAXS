### exploring with the 33mg dataset

#### 0. store csv in array

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from numpy import genfromtxt
from sklearn.decomposition import FastICA, PCA

filename='./data/33_array.csv'
```

```python
### NUMPY OPTION ###
# see https://docs.scipy.org/doc/numpy/reference/generated/numpy.genfromtxt.html
data = genfromtxt(filename, delimiter=',')
data
```

```python
plt.imshow(np.log(data.T), cmap='hot',vmin=0,vmax=10,origin='upper')
plt.show()
```

```python
data.shape
```

#### 1. fastICA

```python
ica = FastICA(n_components=3)
S_ = ica.fit_transform(data)  # Reconstruct signals
A_ = ica.mixing_  # Get estimated mixing matrix
```
